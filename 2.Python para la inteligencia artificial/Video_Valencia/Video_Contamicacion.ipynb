```python
import requests
import pandas as pd
from time import sleep
import os
from os import path
from pathlib import Path
import numpy as np
```

```python
def make_request(base_url, params):
    response = requests.get(base_url, params=params)
    if response.status_code != 200:
        raise Exception(f"Error fetching data: {response.status_code} - {response.text}")
    else:
        return response.json()
```

```python
base_url = "https://valencia.opendatasoft.com/api/explore/v2.1/catalog/datasets/estacions-contaminacio-atmosferiques-estaciones-contaminacion-atmosfericas/records"
```

```python
params = {'limit': 0}
n_stations = make_request(base_url, params)['total_count']
print("Total number of Acctive statrions:", n_stations)
```

```python


res = os.path.join("res")
#print(res)

first_time = True
max_records_pre_request = 100
current_time = 0
sleep_time = 60 * 60
total_time = (24 * 60 * 60) + sleep_time

#to test
total_time = 10
sleep_time = 1

path_csv = [res, 'valencia_pollution_dataset.csv']
path_csv_solved = path.join(*path_csv)
```

```python
while current_time < total_time:
    offset = 0
    pullotion_list = []

    while offset < n_stations:
        params = {
            'limit': max_records_pre_request,
            'offset': offset
        }
        sub_list = make_request(base_url, params)['results']

        pullotion_list += sub_list
        offset += max_records_pre_request

    print(f"Current Time = {current_time}, records processed = {len(pullotion_list)}")

    df = pd.DataFrame(pullotion_list, 
                      columns=["objectid", "nombre", "direccion", "tipozona", "so2", "no2", "o3", "co", "pm10", "pm25","fecha_carg", "calidad_am"])

    df.to_csv(path_csv_solved, sep=',', header=True if first_time else False, index = False, mode='a', na_rep='N/A')
    first_time = False

    sleep(sleep_time)
    current_time += sleep_time
```

```python
# Read the CSV file and display its content
csv_data = pd.read_csv(path_csv_solved, sep=',', header=0, index_col=False, engine='python')

display(csv_data.head(5))
display(csv_data.info()) 
display(csv_data.describe(include='all'))
```

```python
# Filtrar data vacia
filter = csv_data.loc[pd.isnull(csv_data['so2']) | pd.isnull(csv_data['no2']) | pd.isnull(csv_data['o3']) |
                   pd.isnull(csv_data['co']) | pd.isnull(csv_data['pm10']) | pd.isnull(csv_data['pm25'])]

csv_data_filtered = csv_data.copy()
csv_data_filtered.drop(filter.index, inplace=True)

# Convert to datetime
csv_data['fecha_carg'] = pd.to_datetime(csv_data['fecha_carg'])

# Format'YYYY-MM-DD HH:MM:SS'
csv_data['fecha_carg'] = csv_data['fecha_carg'].dt.strftime('%Y-%m-%d %H:%M:%S')

print("Elementos original: ", len(csv_data),"Elementos FIltro: ",  len(filter), "Elementos final: ", len(csv_data_filtered))

# Posibles nombres
#display(csv_data['nombre'].unique().tolist())
#display(csv_data['direccion'].unique().tolist())
#display(csv_data['tipozona'].unique().tolist())

# Mostrar los centros metereológicos únicos
csv_data_uniq_sources = csv_data[['nombre','direccion','tipozona']].drop_duplicates(subset=['nombre','direccion','tipozona'])
#display(csv_data_uniq_sources)

# Mostrar las mediciones maxima de c02
csv_data_max_so2 = csv_data[csv_data['so2'] == csv_data['so2'].max()]
#display(csv_data_max_so2)
display(csv_data_max_so2[['fecha_carg','so2']].drop_duplicates(subset=['fecha_carg']))

#Cuantas veces se ha medido el so2 maximo
print("Mediciones de so2 maximas: ",len(csv_data_max_so2))

# Agrupar por nombre y contar las mediciones de so2 maximo
conteo_so2 = csv_data_max_so2.groupby(['nombre', 'fecha_carg'])['so2'].count().reset_index()
conteo_so2 = conteo_so2.rename(columns={'so2': 'cantidad_mediciones_so2_maximo'})

conteo_so2_con_mediciones = csv_data_max_so2.merge(conteo_so2, on=['nombre', 'fecha_carg'], how='left').drop_duplicates(subset=['nombre','fecha_carg'])


display(conteo_so2_con_mediciones)

```
